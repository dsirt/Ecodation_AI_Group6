```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn-darkgrid') #Tema ayarlamasÄ±
```

```python
x = np.linspace(0, 10, 2000)
plt.figure(figsize = (10, 6))
plt.plot(x, np.sin(x), label = '$Sin(X)$')
plt.plot(x, np.cos(x), label = '$Cos(X)$')
plt.xlabel(r'$X$', fontsize = 18)
plt.ylabel(r'$Y$', fontsize = 18)
plt.title("$Sin(x) $ $ & $ $ Cos(x) $", fontsize = 14)
plt.legend(loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize = (10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x, 2)
y2 = np.power(x, 3)
plt.plot(x, y1, "b-", label = '$y1 = x^2$')
plt.plot(x, y2, color = 'red', linewidth = 1.0, linestyle = '--')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc = 'upper center', fontsize = 'large')
plt.show()
```

```python
x = np.linspace(0, 10, 2000)
plt.figure(figsize = (10, 6))
plt.plot(x, np.sin(x), label = '$Sin(X)$')
plt.plot(x, np.cos(x), label = '$Cos(X)$')
plt.fill_between(x, 0, np.sin(x))
plt.fill_between(x, 0, np.cos(x))
plt.xlabel(r'$X$', fontsize = 18)
plt.ylabel(r'$Y$', fontsize = 18)
plt.title("$Sin(x) $ $ & $ $ Cos(x) $", fontsize = 14)
plt.legend(loc = 'lower left')
plt.show()
```

```python
plt.figure(figsize = (10,6))
x = [1,2,3,4,5,6,7,8,9]
y = [7,8,9,10,7,11,12,13,14]
y2 = [7,7,7,7,7,7,7,7,7]
plt.plot(x, y, linewidth = 4, label = 'Line-1')
plt.plot(x, y2, linewidth = 3, label = 'Line-2')
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title("Line Graph")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,6))
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(1,2,1)
plt.plot(x, y1, color = 'yellowgreen')
plt.subplot(1,2,2)
plt.plot(x, y2, "r")
plt.show()
```

```python
plt.figure(figsize=(12,6))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y1,'b-')
plt.subplot(2,1,2)
plt.plot(x,y2,'r-')
plt.tight_layout()
plt.show()
```

```python
#square, cubic, tangent, exponential
```

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.tan(x)
y4 = np.exp(x)

fig1, ax1 = plt.subplots(nrows = 2, ncols = 2, figsize = (20,20))

ax1[0, 0].plot(x, y1, color = 'blue')
ax1[0, 0].set_title("Square Function")
ax1[0, 0].set_xlabel(r'$X$', fontsize = 18)
ax1[0, 0].set_ylabel(r'$Y$', fontsize = 18)

ax1[0, 1].plot(x, y2, color = 'orange')
ax1[0,1].set_title("KÃ¼p fonksiyon")
ax1[0,1].set_xlabel(r'$x$',fontsize=10)
ax1[0,1].set_xlabel(r'$y$',fontsize=10)

ax1[1,0].plot(x,y3,color="yellow")
ax1[1,0].set_title("Tan fonksiyon")
ax1[1,0].set_xlabel(r'$x$',fontsize=10)
ax1[1,0].set_xlabel(r'$y$',fontsize=10)

ax1[1,1].plot(x,y4,color="brown")
ax1[1,1].set_title("exp fonksiyon")
ax1[1,1].set_xlabel(r'$x$',fontsize=10)
ax1[1,1].set_xlabel(r'$y$',fontsize=10)

plt.show()
```

```python
x1 = [1,3,5,7]
x2 = [2,4,6,8]
y1 = [7,7,7,7]
y2 = [17,18,29,40]
plt.figure(figsize = (8,6))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x1, y1, label = 'First', color = '#42B300')
plt.bar(x2, y2, label = 'Second', color = '#94E413')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Bar Graph")
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
x1 = ["Sayit", "Kerem", "Kubilay", "Ãmer"]
y1 = np.array([17, 18, 29, 40])
y2 = np.array([20, 21, 22, 23])
y3 = np.array([5, 9, 11, 12])
plt.figure(figsize = (5,7))
plt.bar(x1, y1, label = 'SatÄ±Åa AÃ§Ä±k Biletler', width = 0.5, color = '#FF6F00')
plt.bar(x1, y2, label = 'SatÄ±Åa KapalÄ± Biletler', width = 0.5, bottom = y1, color = '#FFB300')
plt.bar(x1, y3, label = 'Ä°ptal Edilen Biletler', width = 0.5, bottom = y1 + y2, color = '#F7DC6F')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Bar Graph")
plt.legend()
plt.show()
```

```python
x1 = np.array([250, 150, 350, 252, 450, 550, 455, 358, 158, 355])
y1 = np.array([40, 50, 80, 90, 100, 50, 60, 88, 54, 45])

x2 = np.array([200, 100, 300, 220, 400, 500, 450, 380, 180, 350])
y2 = np.array([400, 500, 800, 900, 1000, 500, 600, 808, 504, 405])

plt.scatter(x1, y1)
plt.xlabel('$ time $ $ spent $', fontsize = 12)
plt.ylabel('$ score $', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
plt.scatter(x2, y2, color = 'red')
plt.xlabel('$ time $ $ spent $', fontsize = 12)
plt.ylabel('$ score $', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
plt.scatter(x1, y1, label = 'Class 1', marker = 'o', color = 'm')
plt.scatter(x2, y2, label = 'Class 2', marker = 'v', color = 'r')
plt.xlabel('$ time $ $ spent $', fontsize = 12)
plt.ylabel('$ score $', fontsize = 12)
plt.title('Scatter Graph')
plt.legend()
plt.show()
```
