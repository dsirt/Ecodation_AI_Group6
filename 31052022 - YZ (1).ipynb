```python
import numpy as np
from scipy import linalg
A = np.array([[1,2],[3,4]])
print(A)
B = linalg.inv(A)
print(B)
```

```python
A = np.array([[1,3,5],[2,5,1],[2,3,8]])
b = np.array([[10], [8], [3]])
print("Matrix inverse ile denklem Ã§Ã¶zÃ¼mÃ¼ :")
print(linalg.inv(A).dot(b)) #YavaÅ ÃÃ¶zÃ¼m
print("solve() ile denklem Ã§Ã¶zÃ¼mÃ¼ :")
print(np.linalg.solve(A, b)) #HÄ±zlÄ± Ã§Ã¶zÃ¼m
```

```python
from scipy import integrate
x2 = lambda x : x ** 2
I = integrate.quad(x2, 0, 6)
print(I)
```
